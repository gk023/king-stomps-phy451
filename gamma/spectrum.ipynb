```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
energy=[]
counts=[]
error=[]

f=open('137cs-hpge.txt')
for line in f.readlines():
    line = line.strip().split()
    nrg = float(line[0])
    net = float(line[2])
    err = 0.0197*net+np.sqrt(net)
    
    energy.append(nrg)
    counts.append(net)
    error.append(err)
    
f.close()


    
```

```python
plt.rc('text', usetex=True)

plt.figure(figsize=(24,12))
plt.errorbar(energy,counts,yerr=error,fmt='o',ls="none",color="darkviolet",markersize='6')
plt.xlim(0,1000)
plt.yscale('log',basey=16)
plt.xlabel("Channel Number",fontsize=50)
plt.ylabel("Counts",fontsize=50)
plt.tick_params(axis='x',labelsize=50)
plt.tick_params(axis='y',labelsize=50)
plt.savefig('137cs-hpge-nice.png')
```
