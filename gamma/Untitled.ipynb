```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
counts_noshield=[]
energy_noshield=[]

counts_shield=[]
energy_shield=[]


f=open('bkg.txt','r')
for line in f.readlines():
    line = line.strip().split()
    nrg=float(line[0])
    sval=float(line[1])

    energy_shield.append(nrg)
    counts_shield.append(sval)

    
f.close()

g=open('al-bkg.txt','r')
for line in g.readlines():
    line = line.strip().split()
    mrg=float(line[0])
    tval=float(line[1])

    energy_noshield.append(mrg)
    counts_noshield.append(tval)

    
g.close()
```

```python
difference=[]

for i in range(len(counts_shield)):
    diff=counts_shield[i]-counts_noshield[i]
    
    difference.append(diff)
    
print(np.amax(difference))
maxind=difference.index(np.amax(difference))
print(maxind)
print(energy_shield[maxind])
```

```python
plt.rc('text', usetex=True)


fig=plt.figure(figsize=(16,8))
plt.scatter(energy_shield,difference,s=20,marker='o',color='darkblue')
plt.xlabel(r'Energy (keV)',fontsize=35)
plt.ylabel('Counts',fontsize=35)
plt.tick_params(axis='x',labelsize=35)
plt.tick_params(axis='y',labelsize=35)


plt.savefig('removed-bkg-al.pdf')
plt.savefig('removed-bkg-al.png')
plt.show()
```
